**Anonymiser la base de données en gardant la colonne "*visitIp*" et la colonne "*userId*"**

- Anonymiser avec la fonction "*SHA-256*" la colonne "*visitIp*"

- Anonymiser les adresses e-mail dans toute la base de données

- Supprimer la colonne "*idSite*"

```python
import pandas as pd
import hashlib
```

```python
# Chemin fichier
file_path = "~/Desktop/Anonymiser/both/BDDE.csv" # Changer si besoin

# Importation du fichier
df = pd.read_csv(file_path, sep="\t", encoding="utf-16")
# df = pd.read_csv(file_path, sep=",", encoding="utf-16") # Pour M. Prosper

# Premières lignes
df.head()
```

```python
# Fonction pour anonymiser
def anonymise_data(text):
    return hashlib.sha256(text.encode('utf-8')).hexdigest()
```

```python
# Dictionnaire qui associe chaque email à un IP anonymisé
email_to_ip_anonymised = {}
```

```python
def anonymise_columns(row):
    # Assurer que les colonnes sont accessibles 
    user_email = row['userId']
    visit_ip = row['visitIp']
    
    # Anonymiser l'email
    if pd.notna(user_email):  # Si l'email n'est pas vide
        if user_email not in email_to_ip_anonymised:
            email_to_ip_anonymised[user_email] = anonymise_data(visit_ip)  # On associe l'email à l'IP anonymisé
        anonymised_email = anonymise_data(user_email)
        anonymised_ip = email_to_ip_anonymised[user_email]  # Utiliser la même valeur anonymisée pour l'IP
    else:
        anonymised_email = None  # Si pas d'email, on laisse vide
        anonymised_ip = anonymise_data(visit_ip)  # Anonymiser l'IP même si l'email est vide
    
    return pd.Series([anonymised_email, anonymised_ip])
```

```python
# Appliquer l'anonymisation sur les colonnes 'userId' et 'visitIp'
df[['userId', 'visitIp']] = df.apply(anonymise_columns, axis=1)
```

```python
# Sauvegarder le fichier anonymisé
output_path = "~/Desktop/Anonymiser/both/BDDEA_both.csv"
df.to_csv(output_path, sep="\t", encoding="utf-16", index=False)
```
