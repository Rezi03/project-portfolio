# Chargement des librairies

```python
# Librairies de base
import pandas as pd
import numpy as np

# Graphiques
import matplotlib.pyplot as plt
import seaborn as sns

# Clustering
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt
```

# Importer la base de données

```python
file_path = '~/Desktop/projet_la/bdd/bdd.csv'

# Essai avec l'encodage 'latin1'
df = pd.read_csv(file_path, sep=';', encoding='UTF-8-SIG', low_memory=False)

df.head()
```

# Représentation d'un individu par les durées

- *Récupération des identifiants uniques* (`visitorId`)

- *Pour chaque utilisateur :*

   - Calcule le **nombre de visites** en comptant les occurrences de l'identifiant.
     
   - Calcule le **temps moyen** passé par visite
 
   - Calcule l'**écart-type** des temps moyens
 
   - Calcule les **quantiles** (Q1 et Q3) des temps moyens

- *Stocke ces caractéristiques dans une liste et crée un **DataFrame***.

Un **DataFrame** contenant les caractéristiques de chaque utilisateur (Identifiant, temps moyen, écart-type, q1 et q3

```python
# Fonction pour calculer les caractéristiques de chaque utilisateur
def compute_user_features(df):
    # Récupérer les identifiants uniques
    users = df['visitorId'].unique()
    
    # Récupérer les colonnes timeSpent
    time_columns = [col for col in df.columns if col.startswith("timeSpent (actionDetails ")]

    # Liste pour stocker les caractéristiques
    user_features = []

    for user in users:
        # Filtrer les données
        user_data = df[df['visitorId'] == user]

        # Nb de visites
        n_visits = user_data.shape[0]

        # Calcul du tp moyen pour chaque visite (somme des colonnes "timeSpent"/218)
        visit_means = user_data[time_columns].sum(axis=1) / len(time_columns)

        # Calcul de la moyenne des visites
        mean_time = visit_means.mean()

        # Calcul de l'écart-type des temps moyens
        std_time = visit_means.std()

        # Calcul des quantiles (Q1 et Q2/médiane)
        q1 = np.percentile(visit_means, 25)
        q2 = np.percentile(visit_means, 75) 

        # Stocker les résultats
        user_features.append([user, n_visits, mean_time, std_time, q1, q2])

    # Création du DataFrame
    user_features_df = pd.DataFrame(user_features, columns=['visitorId', 'n_visits', 'mean_time', 'std_time', 'q1', 'q2'])

    # Décroissant nombre de visites 
    user_features_df = user_features_df.sort_values(by='n_visits', ascending=False)

    return user_features_df
```

```python
# Afficher la data frame
user_features_df = compute_user_features(df)

print(user_features_df.head(30))
```

```python
# Présence de NaN
nan_counts = user_features_df.isna().sum()

# Afficher
print("Nombre de NaN dans chaque colonne :")
print(nan_counts)

# Vérifier dans toute la DataFrame
if nan_counts.sum() > 0:
    print("\nIl y a des valeurs manquantes (NaN) dans la DataFrame.")
else:
    print("\nIl n'y a pas de valeurs manquantes (NaN) dans la DataFrame.")
```

```python
# Remplacer les NaN dans la colonne 'std_time' par 0
user_features_df['std_time'] = user_features_df['std_time'].fillna(0)

# Vérifier à nouveau s'il y a des NaN
nan_counts = user_features_df.isna().sum()
print(nan_counts)
```

# Clustering hiérarchique ascendant

```python
# Créer un vecteur global par utilisateur (en moyennant les caractéristiques)
def create_user_vectors(user_features_df):
    # Sélectionner les colonnes (moyenne, écart-type, quantiles)
    features = user_features_df[['mean_time', 'std_time', 'q1', 'q2']].values
    
    # Normalisation des données
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)
    
    return features_scaled

# La fonction pour créer les vecteurs
user_vectors = create_user_vectors(user_features_df)

# Le clustering hiérarchique ascendant avec la méthode de Ward (minimise la variance dans chaque cluster)
Z = linkage(user_vectors, method='ward')

# Dendrogramme
plt.figure(figsize=(15, 10))
dendrogram(Z)

# Lables
plt.title("Dendrogramme du Clustering Hiérarchique Ascendant", fontsize=16)
plt.xlabel("ID Utilisateur", fontsize=12)
plt.ylabel("Distance", fontsize=12)

# Rotation des labels
plt.xticks(rotation=90, fontsize=3)
plt.yticks(fontsize=10)

# Ajuster les éléments du graphique
plt.tight_layout()  
plt.show()

# Seuil pour déterminer le nombre de clusters
max_d = 5
clusters = fcluster(Z, max_d, criterion='distance')

# Labels de cluster dans le DataFrame
user_features_df['cluster'] = clusters

# Afficher
print(user_features_df.head())
```

```python
# Afficher la répartition des utilisateurs par cluster
cluster_counts = user_features_df['cluster'].value_counts()

# Afficher les clusters et leur nombre d'individus
print("Répartition des utilisateurs par cluster :")
print(cluster_counts)

# Afficher le cluster avec le plus grand nombre d'individus
max_cluster = cluster_counts.idxmax()
max_cluster_count = cluster_counts.max()
print(f"\nCluster avec le plus d'individus : Cluster {max_cluster} ({max_cluster_count} utilisateurs)")

# Afficher les clusters triés par nombre d'individus
sorted_clusters = cluster_counts.sort_values(ascending=False)
print("\nClusters triés par nombre d'individus :")
print(sorted_clusters)

# Afficher les détails des utilisateurs dans le plus grand cluster
largest_cluster_users = user_features_df[user_features_df['cluster'] == max_cluster]
print("\nDétails des utilisateurs dans le plus grand cluster :")
print(largest_cluster_users)
```
